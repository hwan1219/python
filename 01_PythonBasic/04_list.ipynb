```python
# list01.py
#  
fruits = ['apple', 'banana', 'cherry', 'apple']
print(fruits) # : ['apple', 'banana', 'cherry', 'apple']
```

```python
# list02.py
#  
fruits = ['apple', 'banana', 'cherry']
print(fruits[0]) # : apple
print(fruits[-1]) # : cherry
```

```python
# list03.py
#  
# [:] -    (range   -1)
# [::] #    
fruits = ['apple', 'banana', 'cherry', 'melon', 'mango']
print(fruits[1:4]) # : ['banana', 'cherry', 'melon']
# index() 1  3     

print(fruits[:3]) # : ['apple', 'banana', 'cherry']
# index() 0  2     

print(fruits[::2]) # : ['apple', 'cherry', 'mango']
# :: -  
# 2  0, 2, 4    
```

```python
# list05.py
#   (Nested List)
#        ( )
# 2       (  )
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[0]) # : [1, 2, 3]
print(matrix[1][2]) # : 6
```

```python
# list06.py
#  
#       
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
  print(fruit) #: apple, banana, cherry
```
