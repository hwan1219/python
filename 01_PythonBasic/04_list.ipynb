```python
# list01.py
# 리스트 선언
fruits = ['apple', 'banana', 'cherry', 'apple']
print(fruits) # 출력: ['apple', 'banana', 'cherry', 'apple']
```

```python
# list02.py
# 리스트 인덱싱
fruits = ['apple', 'banana', 'cherry']
print(fruits[0]) # 출력: apple
print(fruits[-1]) # 출력: cherry
```

```python
# list03.py
# 리스트 슬라이싱
# 리스트명[시작:끝] - 끝은 포함되지 않음 (range와 같은 규칙 -1)
# 리스트명[시작:끝:간격] # 간격을 지정하여 요소를 추출
fruits = ['apple', 'banana', 'cherry', 'melon', 'mango']
print(fruits[1:4]) # 출력: ['banana', 'cherry', 'melon']
# index(순번)이 1인 요소부터 3인 요소까지 그 데이터 값을 가져옴

print(fruits[:3]) # 출력: ['apple', 'banana', 'cherry']
# index(순번)이 0인 요소부터 2인 요소까지 그 데이터 값을 가져옴

print(fruits[::2]) # 출력: ['apple', 'cherry', 'mango']
# :: - 스텝을 지정
# 2이므로 인덱스가 0, 2, 4인 요소의 그 데이터를 가져옴
```

```python
# list04.py
# 리스트 주요 메소드
# append(): 리스트의 끝에 새로운 요소를 추가
# insert(): 특정 위치에 요소를 삽입
# remove(): 리스트에서 첫 번째로 일치하는 요소를 제거 (중복된 요소가 있을 시 첫번째를 제거하는 듯)
# pop(): 지정한 인덱스의 요소를 제거하고 그 값을 반환
# sort(): 리스트의 오름차순으로 정렬
# reverse(0): 리스트이 요소 순서를 반대로 뒤집음
# len(): 리스트의 요소 개수를 반환 (자바스크립트: .length)
```

```python
# list05.py
# 리스트의 중첩 (Nested List)
# 리스트는 다른 리스트를 요소로 가질 수 있음 (중첩 리스트)
# 2차원 이상의 데이터 구조를 표현할 때 사용 (예시 검색 필요)
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[0]) # 출력: [1, 2, 3]
print(matrix[1][2]) # 출력: 6
```

```python
# list06.py
# 리스트의 반복문
# 리스트 요소에 접근해 반복문을 통해 순차적으로 처리
fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
  print(fruit) #출력: apple, banana, cherry
```
