```python
# Q1
'''
- 행렬 계산기 구현
목표: 두 개의 2x2 행렬을 정의하고, 행렬 연산(덧셈, 뺄셈, 곱셈, 전치)을 수행
행렬 A: [1,2], [3,4]
행렬 B: [3,4], [7,8]
'''
import numpy as np
A = [[1, 2], [3, 4]]
B = [[5, 6], [7, 8]]

A = np.array(A)
B = np.array(B)
print("A 행렬:")
print(f"{A}\n")
print("B 행렬:")
print(f"{B}\n")

add_result = A + B # 행렬 덧셈 - 같은 위치의 원소끼리 더함
print("add_result:")
print(f"{add_result}\n")

sub_result = A - B # 행렬 뺄셈 - 같은 위치의 원소리끼리 뺌
print("sub_result:")
print(f"{sub_result}\n")

dot_result = np.dot(A, B) # 행렬 곱셈 - 내적(dot product)을 사용한 곱셈 (@ 연산자 사용)
print("dot_result:")
print(f"{dot_result}\n")

transposed_A = A.T # 전치 행렬 - 행과 열을 바꿔주는 연산
print("transposed_A:")
print(transposed_A)
```

```python
# Q2
'''
- 벡터 내적과 코사인 유사도
목표: 두 벡터의 내적과 크기를 이용해 코사인 유사도를 계산
코사인 유사도는 문서 유사도 계산, 검색 엔진 등에 자주 쓰이는 개념
벡터1: [1, 2]
벡터2: [2, 3]
'''
import numpy as np

vec1 = np.array([1, 2])
vec2 = np.array([2, 3])
dot_product = np.dot(vec1, vec2)
# 벡터의 내적 계산: v1[0] * v2[0] + v1[1] * v2[1]

cosine_sim = dot_product / (np.linalg.norm(vec1) * np.linalg.norm(vec2))
# 코사인 유사도: 벡터의 내적 / (벡터1의 크기 * 벡터2의 크기)
print("코사인 유사도:", cosine_sim)
```

```python
# Q3
'''
- 함수 미분 시각화
목표: 함수 f(x) = x^2 와 그 도함수 f'(x)를 시각화하여 미분 개념을 직관적으로 이해
수치 미분(np.gradient)을 통해 도함수를 근사
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np

# x = np.linspace(-5, 5, 1000)
# y = x**2
# dy = 2 * x

# plt.plot(x, y, label="f(x) = x^2")
# plt.plot(x, dy, label="y' = 2x", linestyle='--')
# plt.title("함수와 도함수 시각화")
# plt.legend()
# plt.grid(True)
# plt.show()


# 함수 만들기 (y=x**2)
def f(x):
  return x**2

# -5부터 5까지 100개 포인트 생성
x = np.linspace(-5, 5, 100)

y = f(x) # 원래 함수 값
dy = np.gradient(y, x) # 수치 미분(gradient)을 사용하여 f'(x) 계산

plt.plot(x, y, label="f(x) = x^2")
plt.plot(x, dy, label="y' = 2x", linestyle='--')
plt.title("함수와 도함수 시각화")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Q4
'''
- 정규분포 시뮬레이션
목표: 평균: 0, 표준편차: 1 인 정규분포에서 데이터를 추출(샘플 1000개)하고 히스토그램으로 시각화
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np

# data = np.random.randn(1000) # 평균, 표준편차 각 0, 1로 고정 / size만 기입
data = np.random.normal(0, 1, 1000) # loc(평균), scale(표준편차), size(뽑을 개수) / 3개 전부 입력

plt.hist(data, bins=30, color= 'skyblue', edgecolor= 'black')
plt.title("정규분포 샘플 히스토그램")
plt.grid(True)
plt.show()
```

```python
# Q5
'''
목표: 주사위를 10,000번 던져서 눈금(1~6)의 빈도 분포를 확인
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np

dice_rolls = np.random.randint(1, 7, 10000)
# 1부터 6까지 정수형 난수 10,000개를 생성 후 dice_rolls 배열에 저장, 즉 주사위를 10,000번 던진 결과와 같음

plt.hist(dice_rolls, bins = np.arange(1, 8) - 0.5, rwidth=0.8, color='skyblue')
# bins=np.arange(1, 8) - 5 는 막대의 간격을 0.5~6.5로 설정해 각 bar들이 가운데에 위치하도록 정렬시킴
# rwidth=0.8 은 bar 너비를 80%로 설정하여 bar들의 간격을 약간씩 띄워 시각적으로 보기 좋게 함

plt.xticks(range(1, 7))
# x축 눈금을 1, 2, 3, 4, 5, 6 으로 설정해서 실제 주사위 눈금처럼 보이게 함
plt.title("주사위 시뮬레이션")
plt.grid()
plt.show()
```

```python
# Q6
'''
- 기초 통계량 계산기
목표: 주어진 데이터 리스트에 대해 평균, 중앙값, 분산, 표준편차를 계산
임의의 데이터 리스트: 3, 7, 2, 9, 4
'''
import numpy as np

data = np.array([3, 7, 2, 9, 4])

print("평균:", np.mean(data)) # .mean(평균) 
print("중앙값:", np.median(data)) # .median(중앙값) 
print("분산:", np.var(data)) # .var(분산) 
print("표준편차:", np.std(data)) # .std(표준편차)
```
