```python
# func01.py
# 함수 정의, 호출
# python의 함수 정의는 def 키워드를 사용
# 함수 정의는 이름과 매개변수를 선언하고, 함수가 실행할 코드를 작성

# 정의된 함수는 함수명과 괄호를 사용하여 호출 가능
# 매개변수가 있는 함수는 기본값이 없을 때에 호출 시 해당 매개변수에 값을 반드시 전달해야 함

# 함수 정의
def greet(name):
  return f"Hello, {name}!"

# 함수 호출
result = greet("Alice")
print(result) # 출력: Hello, Alice!
```

```python
# func02.py
# 매개변수와 반환값 없이 함수 정의
# 이런 함수는 주로 특정 작업을 수행하기 위해 사용
def print_hello():
  print("Hello, World!")
print_hello() # 출력: Hello, World!
```

```python
# func03.py
# 매개변수가 있는 함수
# 매개변수에 값을 넣어, 다양한 값을 처리할 수 있음
def add_numbers(a, b):
  return a + b
result = add_numbers(3, 4)
print(result)
```

```python
# func04.py
# 기본값이 있는 매개변수
# 매개변수에 기본값을 설정하면, 인자를 전달하지 않아도 오류가 뜨지 않고 설정해 둔 기본값으로 대체해서 출력
def greet(name="Guest"):
  return f"Hello, {name}!"

# 기본값 사용
print(greet()) # 출력: Hello, Guest!

# 인자를 전달하여 사용
print(greet("Alice")) # 출력: Hello, Alice!
```

```python
# func05.py
# 여러 개의 반환값
# 함수에서 여러 값을 반환할 수 있음, 이 때 반환값은 tuple로 묶여 반환됨
def calculate(x, y):
  sum_result = x + y
  diff_result = x - y
  return sum_result, diff_result
  # 구조 분해 할당, python에선 unpacking이라 표현
  
sum_val, diff_val = calculate(5, 3)
print(sum_val) # 출력: 8
print(diff_val) # 출력: 2
```

```python
# func06.py
# 위치 인자 (Positional Arguments)
# 호출시 매개변수 순서대로 값을 전달
def multiply(a, b):
  return a * b
result = multiply(3, 4) # 위치 인자 방식
# 순서대로 값 전달 / ex) a=3, b=4
# a에 3을 넣길 위해선 반드시 3을 첫번째 인자값에 넣어야함
print(result)
```

```python
# func07.py
# 키워드 인자 (Keyword Arguments)
# 매개변수 이름과 값을 지정하여 전달할 수 있음
def introduce(name, age):
  print(f"My name is {name} and I am {age} years old.")
introduce(age=25, name="Alice") # 키워드 인자 방식
# 매개변수의 이름과 인자값을 같이 적어주기 때문에 순서를 뒤죽박죽 섞어서 적어도 문제되지 않음
```

```python
# func08.py
# 가변 인자 (Variable-Length Arguments)
# 함수에 넘기는 인자의 개수가 불확실 할 때 *args와 **kwargs를 사용하여 가변 인자를 처리할 수 있음
# *args: tuple 형태로 위치 인자를 받음
# **kwargs: dict 형태로 키워드 인자를 받음
def sum_all(*args):
  return sum(args)
print(sum_all(1, 2, 3)) # 출력: 6

def print_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")
print_details(name="Alice", age=25, city="New York")
```

```python
# func09.py
# lambda 함수 정의 - def 를 사용하지 않고 함수를 정의
# 주로 간단한 작업을 수행하는 함수에 사용되고, 코드에서 일회성으로 짧은 함수를 작성할 때 유용하게 쓰임
def add_numbers(x,y): # old
  return x+y
result = add_numbers(3,4)
print(result)

add = lambda x,y: x+y # new
print(add(3,4))
```

```python
# func10.py
# 람다 함수는 list, tuple 등 반복 가능한 객체와 함께 사용될 때 유용
# ex) 리스트의 각 항목에 대해 제곱 값을 구하는 람다 함수
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) # 출력: [1, 4, 9, 16, 25]
```

```python
# func11.py
# 정렬 시 람다 함수 사용
# sorted()나 sort() 함수에서 key 값을 지정할 때 람다 함수를 사용할 수 있음
# ex) 튜플의 첫번째 값으로 정렬하려면 다음과 같이 람다 함수 사용
tuples = [(1, 'one'), (3, 'three'), (2, 'two')]
sorted_tuples = sorted(tuples, key=lambda x: x[0])
print(sorted_tuples) # 출력: [(1, 'one'), (2, 'two'), (3, 'three')]
```

```python
# func12.py
# 조건문과 함께 사용하는 람다 함수
# 람다 함수 안에서 조건문을 사용하여 더 복잡한 로직을 처리할 수 있음
# ex) 짝수인지 홀수인지 판단하는 람다 함수 정의
is_even = lambda x: True if x % 2 == 0 else False
# even: 짝수, odd: 홀수
print(is_even(4)) # 출력: True
print(is_even(5)) # 출력: False
```
