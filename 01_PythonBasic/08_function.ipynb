```python
# func01.py
#  , 
# python   def  
#     ,    

#       
#              

#  
def greet(name):
  return f"Hello, {name}!"

#  
result = greet("Alice")
print(result) # : Hello, Alice!
```

```python
# func02.py
#     
#        
def print_hello():
  print("Hello, World!")
print_hello() # : Hello, World!
```

```python
# func03.py
#   
#   ,     
def add_numbers(a, b):
  return a + b
result = add_numbers(3, 4)
print(result)
```

```python
# func04.py
#   
#   ,           
def greet(name="Guest"):
  return f"Hello, {name}!"

#  
print(greet()) # : Hello, Guest!

#   
print(greet("Alice")) # : Hello, Alice!
```

```python
# func05.py
#   
#      ,    tuple  
def calculate(x, y):
  sum_result = x + y
  diff_result = x - y
  return sum_result, diff_result
  #   , python unpacking 
  
sum_val, diff_val = calculate(5, 3)
print(sum_val) # : 8
print(diff_val) # : 2
```

```python
# func06.py
#   (Positional Arguments)
#     
def multiply(a, b):
  return a * b
result = multiply(3, 4) #   
#    / ex) a=3, b=4
# a 3    3   
print(result)
```

```python
# func07.py
#   (Keyword Arguments)
#       
def introduce(name, age):
  print(f"My name is {name} and I am {age} years old.")
introduce(age=25, name="Alice") #   
#            
```

```python
# func08.py
#   (Variable-Length Arguments)
#        *args **kwargs      
# *args: tuple    
# **kwargs: dict    
def sum_all(*args):
  return sum(args)
print(sum_all(1, 2, 3)) # : 6

def print_details(**kwargs):
  for key, value in kwargs.items():
    print(f"{key}: {value}")
print_details(name="Alice", age=25, city="New York")
```

```python
# func09.py
# lambda   - def     
#      ,        
def add_numbers(x,y): # old
  return x+y
result = add_numbers(3,4)
print(result)

add = lambda x,y: x+y # new
print(add(3,4))
```

```python
# func10.py
#   list, tuple        
# ex)         
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print(squares) # : [1, 4, 9, 16, 25]
```

```python
# func11.py
#     
# sorted() sort()  key        
# ex)         
tuples = [(1, 'one'), (3, 'three'), (2, 'two')]
sorted_tuples = sorted(tuples, key=lambda x: x[0])
print(sorted_tuples) # : [(1, 'one'), (2, 'two'), (3, 'three')]
```

```python
# func12.py
#     
#           
# ex)      
is_even = lambda x: True if x % 2 == 0 else False
# even: , odd: 
print(is_even(4)) # : True
print(is_even(5)) # : False
```
