```python
# Q1
'''
- 영화 별 평점 시각화
다음과 같은 영화들의 평점을 matplotlib 를 활용하여 그래프로 시각화하세요
영화: Interstellar, Inception, Dune, Tenet, Oppenheimer
평점: [9.0, 8.8, 8.1, 7.5, 8.7]
제목, x축 라벨, y축 라벨, y축 범위 설정, y축 점선 그리드
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

movies = ['Interstellar', 'Inception', 'Dune', 'Tenet', 'Oppenheimer']
ratings = [9.0, 8.8, 8.1, 7.5, 8.7]

plt.bar(movies, ratings, color='orange')
plt.title('영화 별 평점 시각화')
plt.xlabel('영화 제목')
plt.ylabel('평점')
plt.ylim(0,10)
plt.grid(True, axis='y', ls='--')
plt.show()
```

```python
# Q2
'''
- 월별 과일 판매량 시각화
과일별 판매량 데이터를 seaborn 으로 그룹 그래프 형식으로 시각화하세요
과일: 사과, 바나나, 오렌지, 포도
1월: [150, 200, 130, 90]
2월: [180, 220, 120, 100]
pandas melt()로 long-format 변환
sns.barplot()으로 월별 그룹 비교
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import seaborn as sns
import pandas as pd

fruits = ['사과', '바나나', '오렌지', '포도']
Jan = [150, 200, 130, 90]
Feb = [180, 220, 120, 100]

df = pd.DataFrame({
  '과일': fruits,
  '1월': Jan,
  '2월': Feb
})
df_melt = pd.melt(df, id_vars='과일', var_name='월', value_name='판매량')

sns.barplot(data=df_melt, x='과일', y='판매량', hue='월')
plt.title('월별 과일 판매량 비교')
plt.show()
```

```python
# Q3
'''
- 시험 점수 분포 시각화
평균 75점, 표준편차 10인 정규분포 데이터를 생성하고 시각화하세요
np.random.normal()로 100개의 점수 생성
plt.hist()로 분포 시각화
x/y축 라벨 및 타이틀 추가가
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import numpy as np

np.random.seed(0)
scores = np.random.normal(75, 10, 100)
# np.random.normal(loc, scale, size)
# 평균, 표준편차, 생성할 난수의 개수(또는 shape)
plt.hist(scores, bins=10, color='lightgreen', edgecolor= 'black')
plt.xlabel('학생')
plt.ylabel('학생 수')
plt.grid(True)
plt.show()

```

```python
# Q4
'''
- 시험 점수 분포 시각화
요일별 방문자 수를 그래프로 시각화하세요
plt.plot() 활용
marker 로 점 표시, 제목/라벨/그리드 포함
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 135, 140, 150, 160, 300, 250]
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
visitors = [120, 135, 140, 150, 160, 300, 250]

plt.plot(days, visitors, 'rs--')
plt.title('요일별 방문자 수')
plt.xlabel('요일')
plt.ylabel('방문자 수')
plt.grid(True)
plt.show()
```

```python
# Q5
'''
- 타이타닉 생존율 분석
타이타닉 데이터셋을 활용하여 선실 등급(class)별 생존자 수를 시각화하세요
sns.countplot() 사용
hue= 'survived' 옵션을 통해 생존 여부 구분
범례 및 라벨 추가
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import seaborn as sns

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x='class', hue='survived')
plt.title('타이타닉 생존자 수 (선실 등급별)')
plt.xlabel('선실 등급')
plt.ylabel('인원 수')
plt.legend(loc='upper right', title='생존 여부', labels=['사망', '생존'])
plt.show()
```

```python
# Q6
'''
- COVID-19 시계열 시각화
12개월간 COVID-19 확진자 수 데이터를 시게열 그래프로 표현하세요
pandas 로 날짜 데이터 생성
plt.plot() 으로 추세 시각화
x 축 라벨 회전, 격자 포함
확진자 수 = [120, 300, 500, 700, 1100, 900, 950, 870, 770, 600, 400, 300]
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import pandas as pd

# date = pd.date_range(start='2023-01-01', periods=12, freq='ME')
# confirmed = [120, 300, 500, 700, 1100, 900, 950, 870, 770, 600, 400, 300]
data = {
  'date': pd.date_range(start='2023-01-01', periods=12, freq='ME'),
  'confirmed': [120, 300, 500, 700, 1100, 900, 950, 870, 770, 600, 400, 300]
}
df = pd.DataFrame(data)

# plt.plot(date, confirmed, marker='o')
plt.plot(df['date'], df['confirmed'], marker='o')
plt.title('월별 COVID-19 확진자 추이')
plt.xlabel('월')
plt.xticks(rotation=45)
plt.ylabel('확진자 수')
plt.grid(True)
plt.show()
```

```python
# Q7
'''
- 펭귄 산점도 분석
펭귄 데이터셋에서 종(species)별 부리 길이와 날개 길이의 관계를 시각화하세요
sns.scatterplot() 사용
hue, style 옵션에 species 설정
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import seaborn as sns

penguins = sns.load_dataset('penguins')

sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', style='species')
plt.show()
```

```python
# Q8
'''
- 상관관계 히트맵
iris 데이터셋의 수치형 변수들 간의 상관관계를 히트맵으로 시각화하세요
seaborn 의 heatmap 사용
annot=True, cmap 옵션 설정
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import seaborn as sns

iris = sns.load_dataset('iris')
corr = iris.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('iris 데이터 상관관계 히트맵')
plt.show()
```

```python
# Q9
'''
- 지역별 매출 파이 차트
서울, 부산, 대구, 광주 지역의 매출을 파이 차트로 시각화하세요
plt.pie() 사용
labels, autopct, startangle 설정 포함함
labels = ['서울', '부산', '대구', '광주']
sales = [400, 200, 150, 100]
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

labels = ['서울', '부산', '대구', '광주']
sales = [400, 200, 150, 100]
colors = [('skyblue', 1), 'orange', 'green', 'red']

plt.pie(sales, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('지역별 매출 비율')
plt.axis('equal')
plt.show()
```

```python
# Q10
'''
- 직군별 연봉 BoxPlot
개발자와 디자이너의 연봉 분포를 boxplot 으로 시각화하세요
seaborn 의 boxplot 사용
가상 연봉 데이터 생성 포함
직군: ['개발자']*50 + ['디자이너']*50
'''
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.family'] = 'Malgun Gothic'
mpl.rcParams['axes.unicode_minus'] = False

import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(1)
data = {
  'job': ['개발자']*50 + ['디자이너']*50,
  'money': np.concatenate([
    np.random.normal(5500, 700, 50),
    np.random.normal(4700, 500, 50)
  ])
}
df = pd.DataFrame(data)

sns.boxplot(data=df, x='job', y='money')
plt.title('직군별 연봉 분포 비교')
plt.xlabel('직군')
plt.ylabel('연봉')
plt.show()
```
