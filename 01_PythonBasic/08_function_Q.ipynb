```python
# Q1
"""
- 간단 챗봇 응답기
사용자의 입력 메시지를 기반으로 간단한 인사/감사 응답을 하는 챗봇을 만들어보세요
사용 기술: 문자열 탐색 (in), 조건문 분기
함수를 밑에 적힌 각 조건에 맞게 응답하도록 정의
"안녕" 이라는 글자가 들어 있으면 "안녕하세요!"
"고마워" 라는 글자가 들어 있으면 "별말씀을요!"
그 외의 글자가 들어 있으면 "잘 모르겠어요."
"""
user_input = input("AI에게 대화를 시도해보세요")
def AIchat(x):
  if "안녕" in x:
    return "안녕하세요!"
  elif "고마워" in x:
    return "별말씀을요!"
  else:
    return "잘 모르겠어요."
response = AIchat(user_input)
print(response)
```

```python
# Q2
"""
- 이름으로 성별 추정하기
이름의 끝 글자를 이용해 성별을 예측하는 간단한 추정기를 만들어보세요.
사용 기술: 문자열 메서드 endswith()
함수를 밑에 적힌 각 조건에 맞게 응답하도록 정의
"희", "숙" 이라는 글자가 들어 있으면 "여성일 가능성이 높음"
"준", "석" 이라는 글자가 들어 있으면 "남성일 가능성이 높음"
그 외의 글자가 들어 있으면 "판단 어려움"
"""
# endswith - 문자열이 특정 문자(또는 문자열)로 끝나는지 확인할 때 쓰임
name = input("이름을 입력하세요")
def genderPrediction(x):
  if x.endswith(("희", "숙")):
    return "여성일 가능성이 높음"
  elif x.endswith(("준", "석")):
    return "남성일 가능성이 높음"
  else:
    return "판단 어려움"
print(genderPrediction(name))
```

```python
# Q3
"""
- 욕설 필터링 함수
텍스트에서 비속어를 탐지하여 ***로 마스킹하는 함수
사용 기술: list와 replace()
함수 안에서 "바보", "멍청이" 라는 비속어가 들어 있으면 "***" 처리
"""
# chat = input("채팅을 입력하세요")
# def badtextfilter(text):
#   badwords = ["바보", "멍청이"]
#   for word in badwords:
#     text = text.replace(word, "***")
#   return text
# print(badtextfilter(chat))

def badtextfilter(text):
  badwords = ["바보", "멍청이"]
  for word in badwords:
    text = text.replace(word, "***")
  return text
print(badtextfilter("넌 정말 바보야"))
print(badtextfilter("멍청이 같아"))
```

```python
# Q4
"""
- 단어 길이 필터링
문장에서 길이가 2 이상인 단어만 추출하는 함수를 만들어보세요
"나 너 우리 모두" 를 넣었을 때
['우리', '모두']
"""
# user_text = "나 너 우리 모두"
# def lengthfilter(text):
#   words = text.split()
#   filtered = []
#   for word in words:
#     if len(word) >= 2:
#       filtered.append(word)
#   return filtered
# print(lengthfilter(user_text))

def lengthfilter(text):
  return [word for word in text.split() if len(word) >= 2]
print(lengthfilter("나 너 우리 모두"))
  
```

```python
# Q5
"""
- 전화번호 마스킹
전화번호에서 뒷자리 4자리를 ****로 가려주는 함수를 만들어보세요
사용 기술: 문자열 슬라이싱
010-1234-****
"""
def phonenumfilter(number):
  return number[:-4] + "****"
print(phonenumfilter("010-1234-5678"))

from IPython.display import Image
Image(filename='image/04_list_slicing.png', width=400)
```

```python
# Q6
"""
- 이메일 도메인 추출기
사용자가 입력한 이메일 주소에서 도메인 부분만 추출해보세요
사용 기술: 문자열 split()
user@example.com
도메인: example.com
"""
def domain(email):
  return email.split("@")[1]
print(f"도메인: {domain("user@example.com")}")

# return email.split("@") - ['user', 'example.com']
# return email.split("@")[1] - example.com
```

```python
# Q7
"""
- 감정 점수 분석기
문장에서 감정 점수를 계산하고 긍정/부정/중립으로 분류하는 함수입니다
사용 기술: 딕셔너리, 문자열 탐색, 조건 분기
# "좋아": 2, "행복": 3, "기쁨": 2
# "슬퍼": -2, "화남": -3, "짜증": -1
"""

# user_input = input("문장을 입력하세요").split()
# def emotion(sentence):
#   good = ['좋아', '행복', '기쁨']
#   bad = ['슬퍼', '화남', '짜증']
#   if any(word in good for word in sentence):
#     return "긍정"
#   elif any(word in bad for word in sentence):
#     return "부정"
#   else:
#     return "중립"
# print(emotion(user_input))


def sentiment_score(text):
  pos = {"좋아": 2, "행복": 3, "기쁨": 2}
  neg = {"슬퍼": -2, "화남": -3, "짜증": -1}

  score = 0
  for word in text.split():
    if word in pos:
      score += pos[word] # 
    elif word in neg:
      score += neg[word]
  if score > 0:
    return "긍정"
  elif score < 0:
    return "부정"
  else:
    return "중립"
print(sentiment_score("좋아 행복해"))
print(sentiment_score("좋아 슬퍼 화남"))
```
